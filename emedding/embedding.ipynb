```python
from openai import OpenAI
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import re
import nltk
import json

nltk.download('punkt')

client = OpenAI(api_key="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")

# 讀取文件內容
with open('/home/shen/Documents/ELTA_Contest/whisper/text_output/output_with_time.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()
    
timestamps = []
texts = []
for line in lines:
    line = line.split(',')
    start_time = line[0]
    end_time = line[1]
    text = line[2]
    timestamps.append((start_time, end_time))
    texts.append(text)
    print(f'{start_time} - {end_time}: {text}')
    
def get_embedding(text, model="text-embedding-3-large"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding

embedding = []
for i, text in enumerate(texts):
    print(f'Processing {i+1}/{len(texts)}')
    embedding.append(get_embedding(text))
    
# 保存嵌入和時間戳到文件中
import json
embeddings_path = '/home/shen/Documents/ELTA_Contest/emedding/embedding/embedding_data.json'
with open(embeddings_path, 'w') as f:
    json.dump({'timestamps': timestamps, 'texts': texts, 'embeddings': embedding}, f)
```
