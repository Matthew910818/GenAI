```python
import firebase_admin
from firebase_admin import credentials, firestore
import json

cred = credentials.Certificate("/home/shen/Documents/ELTA_Contest/firebase/hackathon-4afdf-firebase-adminsdk-iw8pf-3aa04d7f64.json")
firebase_admin.initialize_app(cred)

# 初始化 Firestore
db = firestore.client()



```

```python
# 讀取 JSON 檔案
with open('/home/shen/Documents/ELTA_Contest/firebase/channel_videos.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

for d in data[:10]:
    # 新增資料
    doc_ref = db.collection(u'video_title').document()
    doc_ref.set({
        'video_string' : d['title'],
    })
    print(u'新增資料：', d)
```
