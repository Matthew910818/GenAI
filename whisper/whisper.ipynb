```python
import numpy as np
import torch
import pandas as pd
import whisper

from tqdm.notebook import tqdm


DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# 加載 Whisper 模型
model = whisper.load_model("large")
# wav_path = "/home/shen/Documents/ELTA_Contest/whisper/audio_data/videoplayback.wav"
wav_path = "/home/shen/Documents/ELTA_Contest/whisper/audio_data/2016-17 歐洲冠軍聯賽 - 6⧸4 皇家馬德里 VS 尤文圖斯 (冠軍賽).mp3"
result = model.transcribe(wav_path)

```

```python
import copy

c_result = copy.deepcopy(result)
```

```python
# delete repeated words
def delete_repeated_words(result):
    result = result.split(" ")
    for i in range(len(result)-1):
        if result[i] == result[i+1]:
            result[i] = ""
    result = " ".join(result)
    return result

c_result = delete_repeated_words(c_result['text'])
```

```python
import srt
from datetime import timedelta

def convert_to_srt(transcript):
    subtitles = []
    for i, segment in enumerate(transcript['segments']):
        start = timedelta(seconds=segment['start'])
        end = timedelta(seconds=segment['end'])
        content = segment['text']
        subtitle = srt.Subtitle(index=i+1, start=start, end=end, content=content)
        subtitles.append(subtitle)
    return srt.compose(subtitles)

srt_content = convert_to_srt(result)

with open("output.srt", "w", encoding="utf-8") as f:
    f.write(srt_content)
```
