```python
import subprocess
import json

# 定義頻道 URL
channel_url = "https://www.youtube.com/@zz59743p6/videos"

# 使用 yt-dlp 獲取頻道上的所有影片信息
result = subprocess.run(
    ["yt-dlp", "-j", "--flat-playlist", channel_url],
    capture_output=True,
    text=True
)

# 解析 JSON 結果
videos = [json.loads(line) for line in result.stdout.splitlines()]

# 將結果保存為 JSON 文件
with open("channel_videos.json", "w", encoding="utf-8") as file:
    json.dump(videos, file, ensure_ascii=False, indent=4)

print("影片信息已保存到 channel_videos.json 文件中")
```
